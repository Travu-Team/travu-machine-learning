```python
import pandas as pd
from icrawler.builtin import BingImageCrawler
from icrawler import ImageDownloader
import os
import time
import random

# Custom downloader to intercept URL without downloading
class URLOnlyDownloader(ImageDownloader):
    def download(self, task, default_ext, timeout=5, **kwargs):
        # Simpan URL ke list untuk dikembalikan
        self.image_urls.append(task['file_url'])

# Baca data CSV
df = pd.read_csv("/content/wisata_indonesia.csv")
if 'Image_Path' not in df.columns:
    df['Image_Path'] = None
if 'Image_Status' not in df.columns:
    df['Image_Status'] = None

def create_search_keyword(row):
    return f"{row['nama_wisata']} {row['provinsi']} tempat wisata"

# Loop pencarian URL gambar
for i, row in df[df['Image_Path'].isna()].iterrows():
    keyword = create_search_keyword(row)
    print(f"Mencari gambar untuk: {keyword}")

    try:
        urls = []
        crawler = BingImageCrawler(downloader_cls=URLOnlyDownloader)
        crawler.downloader.image_urls = urls  # Inject list untuk menyimpan hasil
        crawler.crawl(keyword=keyword, max_num=1)

        if urls:
            df.at[i, 'Image_Path'] = urls[0]
            df.at[i, 'Image_Status'] = "Berhasil"
        else:
            df.at[i, 'Image_Status'] = "Gagal: Tidak ada URL gambar"
    except Exception as e:
        df.at[i, 'Image_Status'] = f"Gagal: {str(e)}"

    time.sleep(random.uniform(3, 6))

    if i % 20 == 0:
        df.to_csv('tourism_with_images_result.csv', index=False)
        print(f"Progress disimpan di baris {i}...")

# Simpan hasil akhir
df.to_csv('tourism_with_images_result.csv', index=False)
print("Scraping selesai semua!")
```
