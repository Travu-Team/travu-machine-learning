```python
import pandas as pd
import os
from bing_image_downloader import downloader

df = pd.read_csv("/content/wisata_indonesia_dengan_deskripsi.csv")
output_root = "gambar_wisata"
os.makedirs(output_root, exist_ok=True)

paths = []

for idx, row in df.iterrows():
    nama = row['nama_wisata']

    if pd.isna(nama):
        paths.append("")
        continue

    folder_name = nama.strip().replace(" ", "_").replace("/", "_")
    folder_path = os.path.join(output_root, folder_name)

    try:
        downloader.download(
            query=nama + " Indonesia",
            limit=2,
            output_dir=output_root,
            adult_filter_off=True,
            force_replace=False,
            timeout=60,
            verbose=True
        )
        print(f"✓ Gambar untuk '{nama}' disimpan di: {folder_path}")
        paths.append(folder_path)
    except Exception as e:
        print(f"✗ Gagal download untuk '{nama}': {e}")
        paths.append("")

df['path'] = paths

df.to_csv("wisata_indonesia_final", index=False, encoding='utf-8-sig')
```

```python
df['path_gambar'] = None

for index, row in df.iterrows():
    nama = row['nama_wisata']
    if pd.notna(nama):
        folder_name = nama.strip().replace(" ", "_").replace("/", "_")
        path = os.path.join("gambar_wisata", folder_name)
        df.at[index, 'path_gambar'] = path

print(df[['nama_wisata', 'path_gambar']].head())
df.to_csv("wisata_indonesia_final", index=False)
```
